## Plotting using `ggplot`

```python
# Because I love ggplot
# NOTE: The use of ggplot is for illustrative purposes only
# The current version doesn't work well with recent pandas
# 

from ggplot import *
```

```python
data = df_sum[(df_sum['prot1'] == 'RpoB') & (df_sum['prot2'] == 'RpoB')]
data_ = data.copy()
# This is how we handle symmetry
data_.rename(columns={'res1': 'res2', 'res2': 'res1'}, inplace=True)
data_ = pd.concat([data, data_])
data_.reset_index(drop=True, inplace=True)

p = ggplot(data_, aes(x='res1', y='res2', size='SpecNum')) \
    + geom_point(alpha=0.4, color='#268bd2')

print p
```

```python
data = df_sum.copy()
# This is how we handle symmetry
data_.rename(columns={'res1': 'res2', 'res2': 'res1'}, inplace=True)
data_ = pd.concat([data, data_])
data_.reset_index(drop=True, inplace=True)

p = ggplot(data_, aes(x='res1', y='res2', size='SpecNum')) \
    + geom_point(alpha=0.4, color='#268bd2') \
    + facet_grid('prot1', 'prot2')

print p
```
